```python
import numpy as np
import pandas as pd
from statsmodels.tsa.stattools import adfuller

# Log-transform applicable variables
df['Log_GDP'] = np.log(df['GDP'])
df['Log_CP'] = np.log(df['Cassava Production'])
df['Log_EXR'] = np.log(df['Exchange Rate'])
df['INF'] = df['Inflation Rate']  # Inflation may not need log

# Reusable ADF function with trend
def adf_with_trend(series, name):
    result = {}
    # Level with trend
    level = adfuller(series.dropna(), autolag='AIC', regression='ct')
    result['Level'] = {'ADF Statistic': level[0], 'p-value': level[1]}
    
    # First difference
    diff1 = adfuller(series.diff().dropna(), autolag='AIC', regression='ct')
    result['1st Diff'] = {'ADF Statistic': diff1[0], 'p-value': diff1[1]}
    
    return pd.DataFrame(result).T.round(4)

# Apply on each variable
for var in ['Log_GDP', 'Log_CP', 'Log_EXR', 'INF']:
    print(f"\nADF Test with Trend for {var}")
    print(adf_with_trend(df[var], var))
```

### Recommended Model: ARDL
#### The Autoregressive Distributed Lag (ARDL) model is ideal when you have a mix of I(0) and I(1) variables. It allows for:

* Short-run and long-run dynamics
* Cointegration testing via bounds testing

```python
# Check for NaNs or infinite values
print(df[['Log_GDP', 'Log_CP', 'EXR_Growth']].isna().sum())
print(np.isinf(df[['Log_GDP', 'Log_CP', 'EXR_Growth']]).sum())

# Drop rows with NaNs or infinite values
df_clean = df[['Log_GDP', 'Log_CP', 'EXR_Growth']].replace([np.inf, -np.inf], np.nan).dropna()
```

```python
df_clean = df[['Log_GDP', 'Log_CP', 'EXR_Growth']].dropna()
df_clean.isna().sum()
```

```python
from statsmodels.tsa.ardl import ardl_select_order

# Define dependent and independent variables
y = df_clean['Log_GDP']
X = df_clean[['Log_CP', 'EXR_Growth']]

# Automatically select optimal lags using AIC
selected_model = ardl_select_order(endog=y, exog=X, maxlag=3, maxorder=3, ic='aic')

# Fit the selected model
result = selected_model.model.fit()

# Display results
print(result.summary())
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
import pandas as pd

# Actual and predicted values
y_true = df_clean['Log_GDP'].iloc[result.model._hold_back:]
y_pred = result.fittedvalues

# Metrics
mae = mean_absolute_error(y_true, y_pred)
mse = mean_squared_error(y_true, y_pred)
rmse = np.sqrt(mse)
mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
r2 = r2_score(y_true, y_pred)

# Create a table
metrics_table = pd.DataFrame({
    'Metric': ['MAE', 'MSE', 'RMSE', 'MAPE (%)', 'R Score'],
    'Value': [mae, mse, rmse, mape, r2]
})

# Display
print(metrics_table)
```
